## Analysing the dataset

### Importing libraries

```python
# Libraries to help with reading and manipulating data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

### Defining functions

```python
# Define the function to create boxplot and histogram 
# The parameters are: 
# 1- data: the dataframe
# 2- feature: the atrtribute of the DF to visualise
# 3- figsize: size of the plot, by default height: 12 inches, width: 7 inches
# 4- kde: Enable or nor this property, by default: False
# 5- bins: Insert the number of bins, by default: None
def histogram_boxplot_fn(data, feature, figsize = (12, 7), kde = False, bins = None):
    f2, (ax_box2, ax_hist2) = plt.subplots(nrows = 2,  sharex = True,  gridspec_kw = {"height_ratios": (0.25, 0.75)}, figsize = figsize) # Generate the object instance
    
    sns.boxplot(data = data, x = feature, ax = ax_box2, showmeans = True, color = "salmon")  # Render the boxplot graph boxplot will be created and a star will indicate the mean value of the column
    
    # Verify if the bins parameter is true or false
    if bins:
      sns.histplot(data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, color = "steelblue") #The histogram will take into account the bins parameter
    else: 
      sns.histplot(data = data, x = feature, kde = kde, ax = ax_hist2, color = "steelblue")  #The histogram won't take into account the bins parameter
    
    ax_hist2.axvline(data[feature].mean(), color = "green", linestyle = "dotted", linewidth = 3, label = 'mean: {:5.0f}'.format(data[feature].mean()))  # Add mean to the histogram, including a legend table inside the chart
    ax_hist2.axvline(data[feature].median(), color = "darkorange", linestyle = "dashdot", linewidth = 3, label = 'median: {:5.0f}'.format(data[feature].median()))  # Add median to the histogram, including a legend table inside the chart
```

```python
# Define the function to create countplot with data labels above of each bar
# The parameters are: 
# 1- data: the dataframe
# 2- feature: the atrtribute of the DF to vidualise
# 3- figsize: size of the plot, by default height: 10 inches, width: 7 inches
# 4- order: sort out feature para: by default: None
# 5- hue: split the feature into different colors based on another category
def countplot_fn(data, feature, figsize = (10, 7), order = None, hue = None):
  total = len(data[feature]) # Declare variable for total
  plt.figure(figsize = figsize) # Create the plot with the paramter size
  # Verify if order exists
  if np.all(order):
    # Verify if hue exists
    if hue:
      ax = sns.countplot(data = data, x = feature, order = order, hue = hue); #The plot will take into account the order and the hue
    else:
      ax = sns.countplot(data = data, x = feature, order = order); #The plot will take into account the order only
  else:
    if hue:
      ax = sns.countplot(data = data, x = feature, hue = hue); #The plot will take into account the hue only
    else:
      ax = sns.countplot(data = data, x = feature); #The plot won't take into account the hue only nor rder
  plt.title(feature) # Adding title
  plt.xticks(rotation=90, fontsize=12) # Rotating X label
  # Print count on top of the bar
  for bar in ax.patches: 
    label = "{:.1f}%".format(100 * bar.get_height() / total)
    # ax.annotate(format(bar.get_height(), '.0f'), # Format the number
    ax.annotate(label, # Format the number
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), # Locate the cursor to wirte the data label
                ha = 'center', va = 'center', #Horizontal and vertical alignment
                size = 10, xytext = (0, 8), textcoords='offset points' # Fontsize, starting point to write, type of coordinates
                )
```

### Extracting the dataset

```python
# Loading the excel file
data_train = pd.read_csv(r'C:\Users\Nestor_Arias\OneDrive - EPAM\EPAM Projects\Research Group\CHURN\customer_churn_dataset-training-master.csv')
data_test = pd.read_csv(r'C:\Users\Nestor_Arias\OneDrive - EPAM\EPAM Projects\Research Group\CHURN\customer_churn_dataset-testing-master.csv')
data = pd.concat([data_train, data_test], axis=0)
data.reset_index(drop=True, inplace=True)
# Observing 10 random samples using a seed equals to 1 to make sure always is displaying the same registers
data.sample(n = 10, random_state = 1)
```

## Exploratory Data Analysis

### Univariate Data Analysis

```python
# Getting numeric columns
numericList = data.columns[(data.dtypes == 'float64') | (data.dtypes == 'int64')].to_list()
print(numericList)
```

```python
# Getting string columns
objectList = data.columns[(data.dtypes == 'object') | (data.dtypes == 'category')].to_list()
print(objectList)
```

```python
for column in numericList:
    histogram_boxplot_fn(data, column, kde = True, bins = 20)
    plt.legend(loc = 0);
```

```python
for column in objectList:
    countplot_fn(data = data, feature = column);
```
